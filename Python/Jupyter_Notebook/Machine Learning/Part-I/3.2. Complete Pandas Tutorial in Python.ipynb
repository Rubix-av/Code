Creating a Pandas DataFrame

```python
# @title Importing Boston Dataset
# importing the boston house price data
import pandas as pd
import numpy as np

data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]
# Feature names (based on the original dataset)
feature_names = [
    "CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM", "AGE",
    "DIS", "RAD", "TAX", "PTRATIO", "B", "LSTAT"
]

# Create a dictionary similar to what load_boston used to provide
boston = {
    "data": data,
    "target": target,
    "feature_names": feature_names
}
```

```python
boston_dataset = boston
```

```python
print(boston_dataset)
```

```python
# pandas DataFrame
boston_df = pd.DataFrame(boston_dataset.get('data'), columns=boston_dataset.get('feature_names'))
```
