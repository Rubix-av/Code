Label Encoding:
- converting the labels into numeric form

```python
# importing the dependencies
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

Label Encoding of Breast Cancer Dataset

```python
# loading the data from csv file to pandas dataframe
cancer_data = pd.read_csv('/content/breast_cancer_data.csv')
```

```python
# first 5 rows of the dataframe
cancer_data.head()
```

```python
cancer_data.head()
```

```python
# finding the count of different labels
cancer_data['diagnosis'].value_counts()
```

```python
# load the label encoder function
label_encode = LabelEncoder()
```

```python
labels = label_encode.fit_transform(cancer_data.diagnosis)
```

```python
# appening the labels to the DataFrame
cancer_data['target'] = labels
```

```python
cancer_data.head()
```

0 --> Benign

1 --> Malignant

Benign got 0 as label and Malignant got 1 as label because when we did label encoding on "diagnosis" column, it arranged the values alphabetically and then assigned the values as 0 and 1. In this case, Benign comes before Malignant alphabetically, that's why it got label as 0 and latter got 1

```python
cancer_data['target'].value_counts()
```
