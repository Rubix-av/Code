```python
# load libs
import numpy as np
import pandas as pd
```

```python
# load wine dataset
from sklearn.datasets import load_wine

X, y = load_wine(return_X_y=True, as_frame=True)
```

```python
# print few samples
X.head()
```

```python
# shape of X and y
print(X.shape)
print(y.shape)
```

```python
# split the dataset into train and test split with 70:30 ration
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
print(X.shape, X_train.shape, X_test.shape)
print(y.shape, y_train.shape, y_test.shape)
```

```python
# Train using DecisionTreeClassifier with random_state=1
from sklearn.tree import DecisionTreeClassifier, plot_tree

clf = DecisionTreeClassifier(random_state=1)

clf.fit(X_train, y_train)
```

```python
# get score on training set
clf.score(X_train, y_train)
```

```python
# get score on testing set
clf.score(X_test, y_test)
```

```python
# get the level of depth of the trained decision tree
import matplotlib.pyplot as plt
plt.figure(figsize=(12,8))
plot_tree(clf, filled=True, feature_names=X.columns)
plt.show()
```
