```python
import numpy as np
```

List Vs Numpy - Time Taken

```python
from time import process_time
```

Time taken by a list

```python
python_list = [i for i in range(10000)]

start_time = process_time()
python_list = [i+5 for i in python_list]
end_time = process_time()

print((end_time - start_time)*1000, "milliseconds")
```

```python
np_array = np.array([i for i in range(10000)])

start_time = process_time()
np_array += 5
end_time = process_time()

print((end_time - start_time)*1000, "milliseconds")
```

Numpy Arrays

```python
# list
list1 = [1,2,3,4,5]
print(list1)
type(list1)
```

```python
np_array = np.array([1,2,3,4,5])
print(np_array)
type(np_array)
```

```python
# creating a 1 dim array
a = np.array([1,2,3,4])
a.shape
```

```python
# creating a 2 dim array
b = np.array([(1,2,3,4), (5,6,7,8)])
b.shape
```

```python
c = np.array([(1,2,3,4), (5,6,7,8)], dtype=float)
print(c)
```

Initial placeholders in numpy array

```python
# create a numpy array of Zeros
x = np.zeros((4,5))
print(x)
```

```python
# create a numpy array of Ones
y = np.ones((3,3))
print(y)
```

```python
# array of a particular value
z = np.full((5,4), 5)
print(z)
```

```python
# create an identity matrix
a = np.eye(4, dtype=int)
print(a)
```

```python
# create a numpy array with random values
b = np.random.random((3,4))
print(b)
```

```python
# random integer values array within a specific range
c = np.random.randint(10,100,(3,5))
print(c)
```

```python
# array of evenly spaced values --> specifying the number of values required
d = np.linspace(10,30,6, dtype=int)
print(d)
```

```python
# array of evenly spaced values --> specifying the step
e = np.arange(10,30,6)
print(e)
```

```python
# convert a list to a numpy array
list2 = [10,20,20,20,50]

np_array = np.asarray(list2)
print(np_array)
```

```python
# array dimension
print(c.shape)
```

Mathematical Operations on NumPy array

```python
list1 = [1,2,3,4,5]
list2 = [6,7,8,9,10]
print(list1 + list2) # concatenates two lists
```

```python
a = np.random.randint(0,10,(3,3))
b = np.random.randint(10,20,(3,3))
print(a)
print(b)
```

```python
print(a+b) # adds the elements in case of numpy ndarray
print(a-b)
print(a*b)
print(a/b)
```

```python
a = np.random.randint(0,10,(3,3))
b = np.random.randint(10,20,(3,3))
```

```python
print(np.add(a,b))
print(np.subtract(a,b))
print(np.multiply(a,b))
print(np.divide(a,b))
```

Array Manipulation

```python
array = np.random.randint(0,10,(2,3))
print(array)
print(array.shape)
```

```python
# transpose
trans = np.transpose(array)
print(trans)
print(trans.shape)
```

```python
array = np.random.randint(0,10,(2,3))
print(array)
print(array.shape)
```

```python
trans2 = array.T
print(trans2)
print(trans2.shape)
```

```python
# reshape a array
a = np.random.randint(0,10,(2,3))
print(a)
print(a.shape)
```

```python
b = a.reshape(3,2)
print(b)
print(b.shape)
```
