**Imbalanced Dataset**

A dataset with an unequal class distribution

```python
# importing the dependencies
import numpy as np
import pandas as pd
```

```python
# loading dataset to pandas dataframe
credit_card_data = pd.read_csv('/content/credit_data.csv')
```

```python
credit_card_data.head()
```

```python
credit_card_data.shape
```

```python
credit_card_data.tail()
```

```python
# distribution of the two classes
credit_card_data['Class'].value_counts()
```

This is Highly Imbalanced Dataset

0 --> Legit Transactions

1 --> Fraudulent Transactions



```python
# separating the legit and fraudulent transactions
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
legit.shape, fraud.shape
```

Under-sampling

Building a sample dataset containing similar distribution of Legit and Fraudulent Transactions

Number of Fraudulent Transactions --> 492

```python
legit_sample = legit.sample(n=492)
```

```python
legit.shape, legit_sample.shape, fraud.shape
```

Concatenate the legit_sample and fraud dataset

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```
