```python
# @title Importing Boston Dataset
# importing the boston house price data
import pandas as pd
import numpy as np

data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]
# Feature names (based on the original dataset)
feature_names = [
    "CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM", "AGE",
    "DIS", "RAD", "TAX", "PTRATIO", "B", "LSTAT"
]

# Create a dictionary similar to what load_boston used to provide
boston = {
    "data": data,
    "target": target,
    "feature_names": feature_names
}
```

```python
house = pd.DataFrame(boston.get('data'), columns=boston.get('feature_names'))
house["PRICE"] = boston.get('target')
```

```python
house
```

Distribution Plot

```python
sns.distplot(house["PRICE"])
```

Correlation Matrix :
- Positive Correlation
- Negative Correlation

```python
correlation = house.corr()
```

```python
# constructing a heatmap
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size': 8}, cmap='Blues')
```
